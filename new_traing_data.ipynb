```python
import torch
import matplotlib.pyplot as plt
```

```python
def pertubate_data(data):
    N, length, dim = data.size(0), data.size(1), data.size(2)
    data = data.reshape(N * length * dim)
    for dat in data:
        dat += torch.randn(1)[0]
    return data.reshape(N, length, dim)
```

```python
line = torch.tensor([[0.0], [5.0], [10.0], [15.0]])
line = line.expand(10,4,1)
data = pertubate_data(line)
```

```python
for dat in data:
    plt.plot(dat)
```
